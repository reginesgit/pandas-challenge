```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data_complete_df = pd.DataFrame(school_data_complete)

school_count = len(school_data_complete_df["School ID"].unique())

student_count = len(school_data_complete_df["Student ID"])

total_budget = school_data_complete_df["budget"].sum()

avg_math_score = school_data_complete_df["math_score"].mean()

avg_reading_score = school_data_complete_df["reading_score"].mean()

stu_passed_math = school_data_complete_df.loc[school_data_complete_df['math_score'] >= 70, :]

nr_stu_passed_math = len(stu_passed_math)

percent_stu_passed_math = (nr_stu_passed_math / student_count) *100

stu_passed_reading = school_data_complete_df.loc[school_data_complete_df['reading_score'] >= 70, :]

nr_stu_passed_reading = len(stu_passed_reading)

percent_stu_passed_reading = (nr_stu_passed_reading / student_count) *100

stu_passed_both_math_read = school_data_complete_df.loc[(school_data_complete_df['math_score'] >= 70) 
                                                        & (school_data_complete_df['reading_score'] >= 70), :]
nr_stu_passed_both_math_read = len(stu_passed_both_math_read)
percent_stu_passed_both_math_read = (nr_stu_passed_both_math_read / student_count) *100

district_summary_df = pd.DataFrame({"Total Nr. of Schools": [school_count], 
                                    "Total Nr. of Students": [student_count], 
                                    "Total Budget": [total_budget],
                                    "Average Math Score": [avg_math_score],
                                    "Average Reading Score": [avg_reading_score],
                                    "% of Students who passed Math": [percent_stu_passed_math],
                                    "% of Students who passed Reading": [percent_stu_passed_reading],
                                    "% of Students who passed Math & Reading": [percent_stu_passed_both_math_read] })
district_summary_df
```
